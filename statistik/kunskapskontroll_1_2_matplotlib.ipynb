### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

In Matplotlib, there are two main ways to create plots:

1. The Object-Oriented (OO) Interface:
This approach involves creating Figure and Axes objects explicitly and using their methods to generate plots.
This is considered the recommended approach because it provides more control and flexibility, especially for complex plots with multiple subplots.

2. The Pyplot State-Machine Interface:
This is a more procedural approach where you use plt (the pyplot module) to create and manage the plot. It’s simpler for quick plots and easier for beginners, but it can be less flexible for advanced use cases.
The pyplot interface keeps track of the current figure and axes implicitly, meaning you don’t explicitly need to create Figure and Axes objects.

Recommended Approach: Object-Oriented (OO) Interface
The Object-Oriented approach is preferred, especially for:

Complex plots (e.g., multiple subplots).
Fine-tuned control over plot elements.
Avoiding confusion when working with multiple figures and axes.
The Pyplot State-Machine interface is more straightforward and useful for simple plots or quick visualizations, but for large or complex projects, the OO approach is generally better.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

1. Figure:
The figure is the overall window or canvas that contains the plot(s). It is the top-level container for all plot elements, and a figure can hold multiple subplots or charts.
Think of it as the frame or container for one or more individual plots.

2. Axes:
Axes refers to the actual plotting area where the data is displayed. It contains both the x-axis and y-axis (and possibly a z-axis in 3D plots), along with the data lines, labels, and other plot elements.
A figure can have multiple axes, each representing a subplot.

3. Axis:
An axis is either the x or y (or z in 3D) line that defines the scale and direction of the plot. It handles the tick marks, labels, and limits.
Each axes object has at least one x-axis and one y-axis.

4. Artist:
An artist is any object that can be drawn on the figure. This includes all visual elements, such as lines, text, shapes, and even the figure and axes themselves.
Everything you see on the plot is an artist.

### When plotting in Matplotlib, what is the expected input data type?

When plotting in Matplotlib, the expected input data types are:

Lists or Tuples: Standard Python lists or tuples can be used for plotting data.
NumPy Arrays: Preferred for efficiency and compatibility with numerical data operations.
Pandas Series or DataFrames: Useful for labeled data and direct plotting.
Python Iterables: Any iterable (e.g., range, generators) can be used.
Scalars: For constants like axis limits or simple plot elements.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -4 to 4 using np.linspace
x = np.linspace(-4, 4, 100)
y = x**2  # y = x^2

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add a title and labels
ax.set_title('Plot of y = x^2 (Object-Oriented)')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# First subplot (Scatter plot)
ax1.scatter(random_data_x, random_data_y, alpha=0.5, color='purple', label='Random Data')
ax1.plot(x, y, color='red', label='y = x^2')  # Add y = x^2 curve
ax1.set_title('Scatter Plot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Second subplot (Bar plot)
ax2.bar(names, values, color=['blue', 'green', 'orange', 'yellow', 'lime'])
ax2.set_title('Fruit Data Bar Plot')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Quantity')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
